- Consider these two sets of numbers
X = –1, 0, 1, 2, 3, 4
Y = –3, –1, 1, 3, 5, 7


- If you take one element from set X and one element from set Y, you can create a pair such as (–1, 3).
- Therefore if X is -1 then Y is -3, If X is 3 then Y is 5.
- The formula for this is `y = 2X - 1`
- An example below using neural networks using tensorflow and Keras APIs.

```python
import numpy as np
from keras.api.models import Sequential
from keras.api.layers import Dense
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Dense
import tensorflow as tf


X = [-1, 0, 1, 2, 3, 4]
Y = [-3, -1, 1, 3, 5, 7]

layer_0 = Dense(units=1, input_shape=[1])

# Creating a model with one neuron (layer).
model = Sequential([
    layer_0
])

# Using the Stochastic Gradient Descent optimizer
# and the Mean Squared Error loss function
# A loss function measures how well the model is doing in training.
model.compile(optimizer='sgd', loss='mean_squared_error')

# Training the model
xs = np.array(X, dtype=float)
ys = np.array(Y, dtype=float)
model.fit(xs, ys, epochs=500)

print(model.predict(np.array([10.0])))
print("Here is what I learned: {}".format(layer_0.get_weights()))
```
